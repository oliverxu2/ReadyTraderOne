```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../logs/rounds/2/match_11/match11_score_board.csv", parse_dates=True)
df.tail(6)
```

```python
df.groupby("Team")["ProfitOrLoss"].mean()
```

```python
fig = plt.figure(figsize=(24,8))
teams = df["Team"].unique()
for team in teams:
    d = df.query(f"Team == '{team}'")
    plt.plot(d["Time"], d["ProfitOrLoss"], label=team)
plt.ylabel("Profit or Loss")
plt.legend(loc=2)
```

```python
fig = plt.figure(figsize=(24,8))
for team in teams:
    d = df.query(f"Team == '{team}'")
    plt.plot(d["Time"], d["EtfPosition"], label=team)
plt.ylabel("ETF Position")
plt.legend(loc=3)
```
